```python
import math
import matplotlib.pyplot as plt
import numpy as np
import csv
import os

def divide_half_circle(radius, n):
    if n <= 0:
        print("Number of points should be greater than zero.")
        return

    angle_per_point = math.pi / n
    coordinates = []
    for i in range(n + 1):
        angle = i * angle_per_point
        x = round(radius * math.cos(angle), 3)
        y = round(radius * math.sin(angle), 3)
        coordinates.append((i+1, x, y))

    return coordinates

def plot_points_with_half_circle(coordinates, radius):
    x_values = [coordinate[1] for coordinate in coordinates]
    y_values = [coordinate[2] for coordinate in coordinates]

    fig, ax = plt.subplots()
    ax.scatter(x_values, y_values, color='red')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    theta = np.linspace(0, np.pi, 100)
    x_circle = radius * np.cos(theta)
    y_circle = radius * np.sin(theta)
    ax.plot(x_circle, y_circle, color='blue')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Points on Half Circle')
    ax.grid(True)
    ax.set_aspect('equal')
    plt.show()

def export_to_csv(coordinates, filename):
    headers = ['Point', 'X', 'Y']

    current_dir = os.getcwd()
    filepath = os.path.join(current_dir, filename)

    with open(filepath, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(coordinates)

    print(f"Data has been exported to {filepath} successfully.")

# Example usage
radius = 5.0
elements = 21
circle_coordinates = divide_half_circle(radius, elements)

for i, coordinate in enumerate(circle_coordinates):
    print("Point", coordinate[0], ": x =", coordinate[1], ", y =", coordinate[2])

filename = 'circle_coordinates.csv'
export_to_csv(circle_coordinates, filename)

plot_points_with_half_circle(circle_coordinates, radius)
```

In the above code, the coordinate origin (0,0) is located at the center of the plot. The x-axis extends horizontally, with negative values to the left of the origin and positive values to the right. The y-axis extends vertically, with negative values below the origin and positive values above.

```python
import math
import matplotlib.pyplot as plt
import numpy as np
import csv
import os

def divide_half_circle(radius, span_length, n):
    if n <= 0:
        print("Number of points should be greater than zero.")
        return

    springing_angle = math.acos(span_length/(2*radius))
    centre_angle = math.pi - 2 * springing_angle

    angle_per_point = centre_angle / n
    coordinates = []

    for i in range(n + 1):
        current_angle = i * angle_per_point
        x = round(radius * math.cos(current_angle), 3)
        y = round(radius * math.sin(current_angle), 3)
        coordinates.append((i+1, x, y))

    return coordinates

def plot_points_with_half_circle(coordinates, radius):
    x_values = [coordinate[1] for coordinate in coordinates]
    y_values = [coordinate[2] for coordinate in coordinates]

    fig, ax = plt.subplots()
    ax.scatter(x_values, y_values, color='red')
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    theta = np.linspace(0, np.pi, 100)
    x_circle = radius * np.cos(theta)
    y_circle = radius * np.sin(theta)
    ax.plot(x_circle, y_circle, color='blue')

    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('Points on Half Circle')
    ax.grid(True)
    ax.set_aspect('equal')
    plt.show()

def export_to_csv(coordinates, filename):
    headers = ['Point', 'X', 'Y']

    current_dir = os.getcwd()
    filepath = os.path.join(current_dir, filename)

    with open(filepath, 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(headers)
        writer.writerows(coordinates)

    print(f"Data has been exported to {filepath} successfully.")

# Example usage
radius = 4.83
span_length = 9.26
elements = 21
circle_coordinates = divide_half_circle(radius, span_length, elements)

for i, coordinate in enumerate(circle_coordinates):
    print("Point", coordinate[0], ": x =", coordinate[1], ", y =", coordinate[2])

filename = 'circle_coordinates.csv'
export_to_csv(circle_coordinates, filename)

plot_points_with_half_circle(circle_coordinates, radius)
```
